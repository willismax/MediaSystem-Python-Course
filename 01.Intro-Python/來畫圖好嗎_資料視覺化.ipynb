## Matplotlib可視化
![](https://matplotlib.org/stable/_static/logo2_compressed.svg)

- https://matplotlib.org/stable/index.html

- Matplotlib 是一個強大的 Python 繪圖庫，用於創建廣泛的靜態、動態和交互式圖形。它是基於 NumPy 的多平台數據可視化工具，與 SciPy 和其他科學計算工具包配合使用效果最佳。以下是 Matplotlib 的一些重要特點和應用場景：

- 主要特點：
  - 多平台和多輸出格式支持： Matplotlib 支持數十種後端和輸出類型，這使得它非常靈活，能夠在多種操作系統上運行，並輸出各種所需格式的圖形，包括但不限於 PNG, PDF, SVG, EPS，甚至是互動式圖形。
  - 豐富的繪圖功能： 使用 Matplotlib，用戶可以繪製各種圖形，如折線圖、柱狀圖、散點圖、餅圖等，並能對圖形進行高度自定義，如調整顏色、標籤、圖例等。
  - 與科學計算生態系的整合： Matplotlib 是科學計算和數據分析生態系統中的基礎組件，與 NumPy、SciPy、pandas 等庫緊密整合，是數據分析和科學研究的重要工具。
- 應用場景：
  - 科學研究： 如哈伯太空望遠鏡團隊使用 Matplotlib 來分析天文數據，展示研究成果。
  - 數據分析： 在商業和學術領域廣泛用於數據探索和呈現結果，尤其適用於需要細致呈現數據圖表的情境。
  - 教育和培訓： 在教育領域，Matplotlib 常被用於教授數據可視化的基本概念和技巧。
- 競爭與創新：
  - 與其他工具的比較： 雖然如 R 語言的 ggplot2、基於網頁的 D3.js 和無需編程的 Tableau/PowerBI 提供了更現代化的視覺效果或更用戶友好的操作界面，Matplotlib 在某些方面可能看起來較為傳統。
  - 創新與社群支持： 儘管面臨競爭，Matplotlib 的龐大使用者社群仍然持續推動其發展。新興的套件，如 Seaborn（其設計理念類似於 ggplot2），增強了 Matplotlib 的繪圖功能，提供了更豐富的視覺效果和更好的用戶體驗。
  
Matplotlib 繼續在其強大的社群支持和持續的開發進程中扮演著重要角色，為用戶提供可靠且功能豐富的數據可視化工具。

### 簡單線條圖(折線圖)Line Plots

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```
# 一張圖顯示兩條線：一起寫再plt.show()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x+2, np.cos(x))
```

- 不同顏色的定義方法

```
#plt.plot(x,y , '--b')

plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')  # 使用顏色編碼rgb cmyk
plt.plot(x, np.sin(x-2), color='0.75')  # 0到1的灰階
plt.plot(x, np.sin(x-3), color='#FFDD44')  # 16進位編碼(RRGGBB from 00 to FF)
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))  # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x-5), color='chartreuse')  # 所有HTML支援的顏色名稱
```

- 不同線條樣式定義方法

```
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted');

# 簡易表達方式:
plt.plot(x, x+5, linestyle='-')  # solid
plt.plot(x, x+6, linestyle='--') # dashed
plt.plot(x, x+7, linestyle='-.') # dashdot
plt.plot(x, x+8, linestyle=':')  # dotted
```

- 顏色及線條樣式簡易寫法

```
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

## 散佈圖Scatter Plots

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

#### 方案1:沿用plt.plot()

```
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
plt.plot(x, y+1, 'om')
```

#### 方案2:使用plt.scatter()

```
plt.scatter(x, y, marker='o');
```

- `plt.scatter()`可以做更多的細微操作，如以下四維圖(x軸、y軸、顏色、大小)
- 但大量運算時耗資源請改用`plt.plot()`，因為`plt.plot()`採用複製，`plt.scatter()`為逐一渲染

```
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();  # 顯示顏色尺度
```

- 鳶尾花Iris資料集視覺化
  - x、y為花萼的長度與寬度
  - 點的大喜為花瓣的寬度
  - 顏色為花的特定種類

```
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

print(features)

plt.scatter(
    features[0],
    features[1],
    alpha=0.2,
    s=100*features[3],
    c=iris.target,
    cmap='viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## 直方圖Histograms

- 直方圖histogram：以長條狀的圖表達數量的統計圖型，長條間**有**連在一起(連續資料適用)。

- 長條圖bar chart ：以長條狀的圖表達數量的統計圖型，長條間**沒有**連在一起(非連續資料、名目資料適用)。

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```
plt.hist(data)
```

```
plt.hist(
    data,
    bins=30,
    alpha=0.5,
    histtype='stepfilled',
    color='steelblue',
    edgecolor='none'
    )
```

- `histtype='stepfilled'`, `alpha`屬性結合表示資料分布

```
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

## 圓餅圖Pie charts

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

district = ('Cambridge City', 'East Cambridgeshire', 'Fenland', 'Huntingdonshire',
            'South Cambridgeshire')
population = (123900, 83800, 95300, 169500, 148800)

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red']

# "explode" the 1st slice (Cambridge City)
explode = (0.2, 0.0, 0, 0, 0)

# Produce pie chart. The slices will be ordered and plotted counter-clockwise.
plt.pie(population, explode=explode, labels=district, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.title("2011 population distribution")
```

## 多重子圖表

- `plt.axes()`可建立畫布，子畫布可用list標示位置，所代表的位置為``[left, bottom, width, height]``

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```
ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

- 建立垂直堆疊的axes

```
fig = plt.figure()
ax1 = fig.add_axes(
          [0.1, 0.5, 0.8, 0.4],
          xticklabels=[],
          ylim=(-1.2, 1.2)
          )
ax2 = fig.add_axes(
          [0.1, 0.1, 0.8, 0.4],
          ylim=(-1.2, 1.2)
          )

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

#### plt.subplot()子圖表簡單網格系統

- plt.subplot(列數, 欄數, 第i圖)

```
for i in range(1, 7):
  plt.subplot(2, 3, i)
  plt.text(0.5, 0.5,
  str((2, 3, i)),
  fontsize=18,
  ha='center'
  )
```

- `plt.subplots_adjust()`的`hspace`和`wspace`屬性以指定寬、高間隔，以子圖的尺寸為度。

```
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

#### 以``plt.subplots()``一次準備好整個網格

```
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
  for j in range(3):
    ax[i, j].text(0.5,0.5,str((i,j)), fontsize=18, ha='center')
fig
```

```
# 建立常態分佈資料
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# 使用axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(
    x,
    40,
    histtype='stepfilled',
    orientation='vertical',
    color='gray'
    )
x_hist.invert_yaxis()

y_hist.hist(
    y,
    40,
    histtype='stepfilled',
    orientation='horizontal',
    color='gray')

y_hist.invert_xaxis()
```
